```python
import torch 
from torch import nn
from torch.utils.data import TensorDataset, DataLoader
from torch.nn         import functional as f
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing   import StandardScaler
```

```python
X, y = load_iris(return_X_y=bool(1))
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.25, random_state=1) 
```

```python
xtrain = StandardScaler().fit_transform(xtrain)
xtest  = StandardScaler().fit_transform(xtest)
```

```python
xtrain = torch.tensor(data=xtrain, dtype=torch.float32)
ytrain = torch.tensor(data=ytrain, dtype=torch.float32).reshape(-1, 1)
```

```python
train = TensorDataset(xtrain, ytrain)
```

```python
torch.manual_seed(seed=32)

train = DataLoader(
    dataset=train, batch_size=2, shuffle=True
)
```

```python
class Model(nn.Module):
    def __init__(self) -> None:
        super().__init__()
        self.layer1 =  nn.Linear(
            in_features=4, out_features=16
        )
        self.layer2 = nn.Linear(
            in_features=16, out_features=3
        )
    def forward(self, x):
        x = f.sigmoid(input=self.layer1(x))
        return self.layer2(x)
```

```python
model = Model()
model
```

```python
lr = 0.001 
loss_fn   = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(
    params=model.parameters(), lr=lr
)
```

```python
loss_hist = [0] * 100
accuracy_hist = [0] * 100

for epoch in range(100):
    for xbatch, ybatch in train:
        pred = model(xbatch)
        # pred = torch.argmax(pred, 1).reshape(-1, 1)
  
        ybatch = ybatch.squeeze().long()
        
        loss = loss_fn(pred, ybatch)
        loss.backward()

        optimizer.step()
        optimizer.zero_grad()  

        loss_hist[epoch] += loss.item() * ybatch.size(0)

        pred_classes = torch.argmax(pred, 1)

        correct = (pred_classes == ybatch).float()
        accuracy_hist[epoch] += correct.sum().item()
    
    loss_hist[epoch] /= len(train.dataset)
    accuracy_hist[epoch] /= len(train.dataset)
    print(loss.item() ,accuracy_hist[epoch])
```

```python
plt.plot(loss_hist)
```

```python
plt.plot(accuracy_hist)
```
