Implementação do perceptron no python

```python
import numpy as np
import matplotlib.pyplot as plt
from implementacao import Perceptron, fronteira_de_decisao, AdalineGD, AdalineSGD
```

```python
from sklearn.datasets import load_iris
import pandas as pd
data = load_iris(as_frame=True)
data['target_names']
```

```python
X = data['data']
y = data['target']
```

```python
y = np.where(y==0, 0, 1)
X = X.iloc[0:150, [0, 2]].values
```

```python
plt.scatter(X[:50, 0], X[:50, 1], label='setosa')
plt.scatter(X[50:100, 0], X[50:100, 1], label='versicolor')
plt.legend()
plt.show()
```

```python
pnn = Perceptron(eta=0.1, n_iter=10)
pnn.fit(X, y)
```

```python
plt.plot([i for i in range(1, len(pnn.errors_)+1)], pnn.errors_, marker='o')
plt.xlabel('Épocas')
plt.ylabel('Número de atualizações')
plt.show()
```

```python
fronteira_de_decisao(X, y, classifier=pnn, color='dark6')
plt.show()
```

```python
adaline = AdalineGD(verbose=True)
adaline.fit(X, y)
```

```python
fronteira_de_decisao(X, y, classifier=adaline, color='dark6')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

X = StandardScaler().fit_transform(X)
```

```python
adaline = AdalineGD(eta=0.1, n_iter=100,verbose=True)
adaline.fit(X, y)
```

```python
fronteira_de_decisao(X, y, classifier=adaline, color='dark6')
plt.show()
```

```python
from sklearn.metrics import classification_report

print(classification_report(y, adaline.predict(X)))
```
