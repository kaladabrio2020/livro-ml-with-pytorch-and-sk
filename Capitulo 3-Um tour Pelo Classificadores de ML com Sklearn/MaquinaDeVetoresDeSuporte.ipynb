```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification, load_iris
from sklearn.svm      import SVC
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import train_test_split
from plots.PlotRegiaDeDecisao import plotDecisionRegion2,plotDecisionRegion
```

```python
Xk , yk = load_iris(return_X_y=True)
yk = np.where(yk == 0,1,0)
Xk = Xk[:,[2,3]]
```

```python
xtraink ,xtestk ,ytraink ,ytestk = train_test_split(Xk, yk, random_state=42, test_size=0.2)
```

### SVM linear

```python
svnmlinear = SVC(C=1, kernel='linear')
svnmlinear.fit(xtraink,ytraink)
```

```python
plotDecisionRegion2(xtraink,ytraink,svnmlinear)
```

#### Usando o gradiente de descendente com a funcao de perda 'hinge'

```python
C = 2
grad = SGDClassifier(loss='hinge',alpha=1/np.size(xtraink,axis=0)*C)
grad.fit(xtraink,ytraink)
```

```python
plotDecisionRegion2(xtraink,ytraink,grad)
```
