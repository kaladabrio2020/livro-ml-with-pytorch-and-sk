```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets     import make_classification
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import StandardScaler
```

### Regularização L2
Ele adciona $\alpha\sum^{n}_{i=n} w_i^2$ a função de perda dependendo do modelo que estiver usando , reduz os valores do peso o menor possivel

```python
X, y = make_classification(n_samples=150,n_classes=2)
X = X[:,[0,1]]
X = StandardScaler().fit_transform(X)
```

```python
x1 = []
x2 = []
for alphai in np.arange(0.1,5,0.1):
    gradiente = SGDClassifier(loss='log_loss',penalty='l2',alpha=alphai)
    gradiente.fit(X,y)

    x1.append(gradiente.coef_[0][0])
    x2.append(gradiente.coef_[0][1])
```

```python
fig , ax = plt.subplots(1,2,figsize=(10,5))
index    = 0
for xi ,axi in zip([x1,x2],ax.flatten()):
   axi.plot(np.arange(0.1,5,0.1),xi)
   axi.set_title(f'caracteristica X{index}')
   index+=1

plt.show()
```

Os pesos das features se aproxima do zero a medida que o $\alpha$ aumanet

### Regularização l1
Ele adiciona $\alpha\sum^{n}_{i=n} |w_i|$  a função de perda, reduz a zero os peso das caracteristicas menos importante

```python
x1 = []
x2 = []
for alphai in np.arange(0.1,5,0.1):
    gradiente = SGDClassifier(loss='log_loss',penalty='l1',alpha=alphai)
    gradiente.fit(X,y)

    x1.append(gradiente.coef_[0][0])
    x2.append(gradiente.coef_[0][1])
```

```python
fig , ax = plt.subplots(1,2,figsize=(10,5))
index    = 0
for xi ,axi in zip([x1,x2],ax.flatten()):
   axi.plot(np.arange(0.1,5,0.1),xi)
   axi.set_title(f'caracteristica X{index}')
   index+=1

plt.show()
```

> Se for muito alto o valor de $\alpha$, os peso das features que não são menos importantes se aproxima de zero
